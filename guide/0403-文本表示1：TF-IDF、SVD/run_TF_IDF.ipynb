```python
import pandas as pd
import sklearn as sk
import math 
```

```python
sentence1 = "love is like pi natural irrational and very important"
sentence2 = "love is being stupid together"
sentence3 = "love is sharing your popcorn"
sentence4 = "love is like heaven but it can hurt like hell"

#split so each word have their own string
sentence1 = sentence1.split(" ")
sentence2 = sentence2.split(" ")
sentence3 = sentence3.split(" ")
sentence4 = sentence4.split(" ")

#join them to remove common duplicate words
total= set(sentence1)
total = total.union(set(sentence2))
total = total.union(set(sentence3))
total = total.union(set(sentence4))

print(total)
```

```python
import nltk
from nltk.corpus import stopwords
stop_list = set(stopwords.words('english'))
#stop_list
```

```python
filtered_total = []

for word in total:
    if str(word) not in set(stopwords.words('english')):
        filtered_total.append(word)
total = filtered_total
total
```

```python
wordDict1 = dict.fromkeys(total, 0) 
wordDict2 = dict.fromkeys(total, 0)
wordDict3 = dict.fromkeys(total, 0)
wordDict4 = dict.fromkeys(total, 0)

for word in sentence1:
    if word in total:
        wordDict1[word]+=1
    
for word in sentence2:
    if word in total:
        wordDict2[word]+=1

for word in sentence3:
    if word in total:
        wordDict3[word]+=1

for word in sentence4:
    if word in total:
        wordDict4[word]+=1
```

```python
def computeTF(wordDict, this_doc):
    tfDict = {}
    doc_len = len(this_doc)
    for word, count in wordDict.items():
        tfDict[word] = count/float(doc_len)
    return(tfDict)
```

```python
#running our sentences through the tf function:
tf1 = computeTF(wordDict1, sentence1)
tf2 = computeTF(wordDict2, sentence2)
tf3 = computeTF(wordDict3, sentence3)
tf4 = computeTF(wordDict4, sentence4)

#Converting to dataframe for visualization
tf = pd.DataFrame([tf1, tf2,tf3,tf4])
tf
```

```python
def computeIDF(docList):
    idfDict = {}
    N = len(docList)
    
    idfDict = dict.fromkeys(docList[0].keys(), 0)
    # print(idfDict)
    for word, val in idfDict.items():
        for doc in docList:
            if doc[word] >=1 :
                idfDict[word] = idfDict[word]+1
            
    print(idfDict)
    for word, val in idfDict.items():
        # idfDict[word] = math.log10(N/ (float(val) + 1))
        if( val ==0 ):
            idfDict[word] = math.log10(N/ (float(val) + 1))
        else:
            idfDict[word] = math.log10(N/ (float(val)))
            # 此处没有不在文档集出现过的query word，所以分母不加1
            # 如果加1，love单词的idf为log(4/5)为一个负数
    return(idfDict)
```

```python
idfs = computeIDF([wordDict1, wordDict2,wordDict3,wordDict4])
idfs
```

```python
def computeTFIDF(tfBow, idfs):
    tfidf = {}
    for word, val in tfBow.items():
        tfidf[word] = val*idfs[word]
    return(tfidf)

#running our sentences through the IDF:
tfidf1 = computeTFIDF(tf1, idfs)
tfidf2 = computeTFIDF(tf2, idfs)
tfidf3 = computeTFIDF(tf3, idfs)
tfidf4 = computeTFIDF(tf4, idfs)

#putting it in a dataframe
tfidf= pd.DataFrame([tfidf1, tfidf2,tfidf3,tfidf4 ])
tfidf
```
